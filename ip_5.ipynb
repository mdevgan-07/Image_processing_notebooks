## **WAP to enhance the input image by performing log transform**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread('/content/grayscale_img.jpg',cv2.IMREAD_UNCHANGED)

img_float=img.astype(np.float32)

c=255/(np.log(1+np.max(img_float)))

log_transformed=c*np.log(1+img_float)
log_transformed=np.uint8(np.clip(log_transformed,0,255))

cv2_imshow(img)
cv2_imshow(log_transformed)

```
