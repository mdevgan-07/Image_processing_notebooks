### **WAP to enhance the input image using gamma transformation**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread('/content/grayscale_img.jpg',cv2.IMREAD_UNCHANGED)

c = 1

gamma = float(input("Enter gamma value: "))

gamma_transformed = c * np.power(img.astype(np.float32)/255.0, gamma)
gamma_transformed = cv2.normalize(gamma_transformed, None, 0, 255,
cv2.NORM_MINMAX).astype(np.uint8)

cv2_imshow(img)
cv2_imshow(gamma_transformed)


```
