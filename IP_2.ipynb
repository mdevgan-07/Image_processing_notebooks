**WAP to convert RGB image to grayscale:**

```python
'''case 1: By taking mean of the three planes'''

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread('/content/rgb_image.jpg')
b,g,r=cv2.split(img)

gray_mean=(r.astype(np.float32) + g.astype(np.float32) + b.astype(np.float32)) / 3
gray_mean=np.uint8(gray_mean)

cv2_imshow(img)
cv2_imshow(gray_mean)
```

```python
'''case 2: Input the weightage of three planes, i.e., R, G, B from the user. The weightage is a value
between 0 and 1, and sum of all three weightage is equal to 1. '''

cv2.imread('/content/rgb_image.jpg')
print("Enter the weightages of the R,G and B respectively(between 0 and 1 and sum=1): ")

w_r=float(input("Enter the value of Red: "))
w_g=float(input("Enter the value of Green: "))
w_b=float(input("Enter the value of Blue: "))

gray_weighted=(w_r*r+w_g*g+w_b*b).astype(np.uint8)

cv2_imshow(img)
cv2_imshow(gray_weighted)
```
