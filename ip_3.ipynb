### **WAP to draw a border around the input image(or in other words pad the input image)**

```python
import cv2
from google.colab.patches import cv2_imshow

img=cv2.imread('/content/rgb_image.jpg')

top=int(input("Enter top border width in pixels: "))
bottom=int(input("Enter the bottom border width in pixels: "))
left=int(input("Enter the left border width in pixels: "))
right=int(input("Enter the right border width in pixels: "))

pink_color=[203,192,255]

bordered_img=cv2.copyMakeBorder(img,top,bottom,left,right,cv2.BORDER_CONSTANT,value=pink_color)

cv2_imshow(img)
cv2_imshow(bordered_img)
```
